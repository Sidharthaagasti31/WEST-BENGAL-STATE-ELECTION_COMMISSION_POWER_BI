```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import Select
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import NoSuchElementException,StaleElementReferenceException
import time
import pandas as pd
import numpy as np
```

```python
# except first page of all dropdownlist
driver = webdriver.Edge(r"C:\Users\Dell\Downloads\edgedriver_win64 (2)\msedgedriver.exe")
driver.get(r"https://portal.wbsec.org/")
dropdown = Select(driver.find_element_by_xpath('//*[@id="cmbDistrict"]'))
all_dropdown = dropdown.options
dropdown_list=[]
for i in all_dropdown:
    dropdown_list.append(i.get_attribute('value'))
    
dataframe=[]
for zillaparishad in dropdown_list:
    dropdown.select_by_value(zillaparishad)
    
    wait = WebDriverWait(driver, 10)
    loader_element = wait.until(EC.invisibility_of_element_located((By.ID, "loader-wrapper")))
    search = driver.find_element_by_xpath('//*[@id="nav-tabContent"]/div[1]/div/section/div/div/div/div/div[2]/div[5]/button')
    search.click()
#     rows = driver.find_elements_by_xpath("//tr[contains(@class, 'ng-scope even')]")
#     for x in rows:
#         print(x.find_element_by_class_name('Cname').text)
#         print(x.find_element_by_css_selector('span[title="Father / Husband"].ng-binding').text)
#         print(x.find_element_by_css_selector('td.text-center.ng-binding[title="Gender"]').text)
#         print(x.find_element_by_css_selector('td.text-left.ng-binding[title="Address"]').text)
#         print(x.find_element_by_css_selector('td.text-center.ng-binding[title="Party Affiliation"]').text)
#         print(x.find_element_by_css_selector('td.text-left.ng-binding[title="Category"]').text)
    while True:
        rows = driver.find_elements_by_css_selector('tr[data-ng-repeat="a in ContestingCandidatesAcceptedPortalListAll"]')
        for i in rows:
            try:
                name = i.find_element_by_class_name('Cname').text
                father = i.find_element_by_css_selector('span[title="Father / Husband"].ng-binding').text
                gender = i.find_element_by_css_selector('td.text-center.ng-binding[title="Gender"]').text
                address = i.find_element_by_css_selector('td.text-left.ng-binding[title="Address"]').text
                party = i.find_element_by_css_selector('td.text-center.ng-binding[title="Party Affiliation"]').text
                category = i.find_element_by_css_selector('td.text-left.ng-binding[title="Category"]').text
                dataframe.append({'name':name,'father/husband':father,'gender':gender,'address':address,'party':party,'category':category})
            except StaleElementReferenceException:
                continue
        try:
            next_button = wait.until(EC.element_to_be_clickable((By.XPATH, '//*[@id="tblContestingCandidatesAcceptedPortalList_next"]/a')))
            next_button.click()
        except:
            break    

data = pd.DataFrame(dataframe)
                                 
data
data.to_csv('WEST_ENGAL_STATE_ELECTION_COMMISSION.csv',index=False)

                                
```
