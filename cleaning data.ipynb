### Cleaning data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df1=pd.read_csv('WEST_ENGAL_STATE_ELECTION_COMMISSION2.csv')
df2=pd.read_csv('WEST_ENGAL_STATE_ELECTION_COMMISSION.csv')
```

```python
df=pd.concat([df1,df2],ignore_index=True)
```

```python
df.isna().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.isna().sum()
```
